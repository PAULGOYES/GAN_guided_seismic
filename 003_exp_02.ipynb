## Utils

```python
def randomsamling(N=28,cr=0.1):
  '''
  cr = compression ratio
  N  = full element number
  '''

  x0 = np.ones((N,))
  ss = np.random.permutation(list(range(1, N - 1)))
  x0[ss[0:int(cr * N)]] = 0

  return x0
```

```python
def subsampling(x,cr=[0.7]):
  '''
  x : full data
  H : random vector 1xN zeros-ones from binomial distribution
  '''
  batch,M,N,L = x.shape

  out = []
  train = []

  for j in cr:
    new = np.zeros_like(x)

    for i in range(batch):
       H = randomsamling(N=N,cr=j)
       new[i,:,:,0] = (x[i,:,:,0]+ (0.5+j)*np.random.normal(0, 0.008, size=(28,28))) * np.tile(H.reshape(1,-1),(x.shape[1],1))
    out.append(new)
    train.append(x)

  return np.concatenate(out,axis=0),np.concatenate(train,axis=0)

from functools import wraps
from time import time
import timeit

def timeitfc(func):
    @wraps(func)
    def timeit_wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        total_time = end_time - start_time
        print(f'Function {func.__name__}{args} {kwargs} Took {total_time:.4f} seconds')
        return result
    return timeit_wrapper
```

```python
def log10(x):
  numerator = K.log(x)
  denominator = K.log(K.constant(10, dtype=numerator.dtype))
  return numerator / denominator

def PSNR(y_true, y_pred):
  msel = tf.keras.losses.MeanSquaredError()
  max_pixel=1
  return 10.0 * log10((max_pixel) / msel(y_true,y_pred))
```

```python
import random

def select_random_indexes(N, M, k, seed=None):
    if M > N - 8:
        raise ValueError("M should be less than or equal to N - 8.")

    # Set the random seed for reproducibility
    if seed is not None:
        random.seed(seed)

    # Generate a random permutation of N indexes excluding the first and last 4
    available_indexes = list(range(4, N - 4))
    random.shuffle(available_indexes)

    # Select the first M indexes from the available indexes
    selected_indexes = available_indexes[:M]

    # Check for consecutive sequences of length k
    while any(selected_indexes[i] - selected_indexes[i - 1] == 1 for i in range(1, M)):
        # If consecutive sequence found, regenerate and try again
        random.shuffle(available_indexes)
        selected_indexes = available_indexes[:M]

    # Add 4 to each selected index to account for the exclusion of the first 4 elements
    selected_indexes = [index + 4 for index in selected_indexes]

    return selected_indexes

# Example usage with a seed for reproducibility
N = 100  # Total number of indexes
M = 20   # Number of random indexes to select
k = 2   # Avoid consecutive sequences of length k
seed = 42  # Random seed for reproducibility

random_indexes = select_random_indexes(N, M, k, seed)
print(np.sort(random_indexes))
```

```python
blksz = (28,28)
stpsz = (1,1)
blcks_corrupt = array.extract_blocks(singleffid, blksz, stpsz)
blcks_non_corrupt = array.extract_blocks(TestData, blksz, stpsz)
```

```python
# bloques en arreglo para ingresar a la red
subblcks_corrupt = blcks_corrupt.transpose(2,0,1)[:,:,:,np.newaxis]
subblcks_non_corrupt = blcks_non_corrupt.transpose(2,0,1)[:,:,:,np.newaxis]

#np.savez('testing1.npz',x=subblcks_corrupt,y=subblcks_non_corrupt)
```

```python
#blques salida de la red - reconstruidos
recov_blcks = modelcombinado.predict(subblcks_corrupt)
```

```python
Urec  = recov_blcks.squeeze().transpose(1,2,0)
Urec = Urec.reshape(np.product(blksz), -1)
```

```python
start = timeit.default_timer()
#imgd_median = array.average_blocks(Urec.reshape(blksz + (-1,)), singleffid.shape, stpsz)
imgd_median = array.combine_blocks(Urec.reshape(blksz + (-1,)), singleffid.shape, stpsz, np.median)
print("creating image from patching, time is:",
              timeit.default_timer() - start)
```
